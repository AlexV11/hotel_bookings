```python
from sklearn.preprocessing import LabelBinarizer
booking.drop('Booking_ID', axis = 1, inplace = True)
lb = LabelBinarizer()
booking['booking_status'] = lb.fit_transform(booking['booking_status'])
booking.head()
```

```python
corr_matrix = booking.copy()
corr_matrix = corr_matrix.corr()
corr_matrix['booking_status'].sort_values(ascending = False)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(booking, booking['booking_status']):
    strat_train_set = booking.loc[train_index]
    strat_test_set = booking.loc[test_index]
```

```python
booking = strat_train_set.drop('booking_status', axis=1)
booking_labels = strat_train_set['booking_status'].copy()
```

```python
booking_num = booking.drop(['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'], axis=1)
booking_num
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

onehot_meal_plan = OneHotEncoder()
onehot_room_type = OneHotEncoder()
onehot_market_segment = OneHotEncoder()

meal_plan_1hot = onehot_meal_plan.fit_transform(booking['type_of_meal_plan'].values.reshape(-1,1))
room_type_1hot = onehot_room_type.fit_transform(booking['room_type_reserved'].values.reshape(-1,1))
market_segment_1hot = onehot_market_segment.fit_transform(booking['market_segment_type'].values.reshape(-1,1))

booking_cat = pd.DataFrame(meal_plan_1hot.toarray())
booking_cat = pd.concat([booking_cat, pd.DataFrame(room_type_1hot.toarray())], axis=1)
booking_cat = pd.concat([booking_cat, pd.DataFrame(market_segment_1hot.toarray())], axis=1)
booking_cat
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import FeatureUnion

num_attribs = list(booking_num)
cat_attribs = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']

num_pipeline_svm = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('std_scaler', StandardScaler())
])

num_pipeline_trees = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('pca', PCA())
])

cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),
    ('one_hot_encoder', OneHotEncoder())
])

full_pipeline_svm = FeatureUnion(transformer_list = [
    ('num_pipeline_svm', num_pipeline_svm),
    ('cat_pipeline', cat_pipeline)
])

full_pipeline_trees = FeatureUnion(transformer_list = [
    ('num_pipeline_trees', num_pipeline_trees),
    ('cat_pipeline', cat_pipeline)
])
```

Evaluate the models on the test set

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, auc, roc_curve, PrecisionRecallDisplay

def print_metrics(y_true, y_pred):
    cf_matrix = confusion_matrix(y_true, y_pred)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')

    print("Accuracy: %.2f%%" % (accuracy_score(y_true, y_pred) * 100.0))
    print("F1 Score: %.2f%%" % (f1_score(y_true, y_pred) * 100.0))
    print("Precision: %.2f%%" % (precision_score(y_true, y_pred) * 100.0))
    print("Recall: %.2f%%" % (recall_score(y_true, y_pred) * 100.0))
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    print("AUC: %.2f%%" % (auc(fpr, tpr) * 100.0))
```
